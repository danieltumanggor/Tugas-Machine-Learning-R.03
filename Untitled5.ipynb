```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/BL-Flickr-Images-Book.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=[
    'Corporate Author',
    'Corporate Contributors',
    'Former owner',
    'Engraver'
], inplace=True)
```

```python
df.head()
```

```python
df['Date of Publication'] = df['Date of Publication'].astype(str)
```

```python
df['Date of Publication'] = df['Date of Publication'].str.extract('(\d{4})')
```

```python
df['Date of Publication'] = pd.to_numeric(df['Date of Publication'])
```

```python
df['Date of Publication'].head(10)
```

```python
df['Author'].value_counts().head(10)
```

```python
df['Place of Publication'].value_counts().head(10)
```

```python
df['Date of Publication'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))

df['Date of Publication'].dropna().astype(int).hist(bins=30)

plt.title('Distribusi Tahun Publikasi Buku')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Buku')

plt.show()
```

```python
top_places = df['Place of Publication'].value_counts().head(10)

top_places.plot(kind='bar', figsize=(10,5))

plt.title('10 Kota Penerbit Terbanyak')
plt.xlabel('Kota')
plt.ylabel('Jumlah Buku')

plt.show()
```
